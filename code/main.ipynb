```python
def format_storage(storage):
    storage = storage.strip()
    
    pattern = r'(\d+)(GB|TB)'
    matches = re.findall(pattern, storage)
    
    formatted_storage = ' '.join([f"{match[0]}{match[1]}" for match in matches])
    
    return formatted_storage

df['Storage'] = df['Storage'].apply(format_storage)
```
